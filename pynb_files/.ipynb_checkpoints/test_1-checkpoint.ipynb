```python
import numpy as np
import matplotlib as plt
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
df = pd.read_csv('../data/Covid Data.csv')
```

```python
df.shape
```

```python
df.head(40)
```

```python
df.columns
```

```python
df.head()
```

```python
df['covid_status_binary'] = df['CLASIFFICATION_FINAL'].apply(
    lambda x: 1 if x in [1, 2, 3] else 0
)
```

```python
df.drop(['INTUBED','PATIENT_TYPE','ICU','CLASIFFICATION_FINAL'], axis=1, inplace=True, errors='ignore')
```

```python
df['died'] = df['DATE_DIED'].apply(lambda x: 0 if x == '9999-99-99' else 1)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(100)
```

```python
cols_to_fix = ['PREGNANT']
df[cols_to_fix] = df[cols_to_fix].replace(97, 0)
df.shape
```

```python
y = df['died']
x = df.drop(['covid_status_binary','USMER','MEDICAL_UNIT','DATE_DIED'],axis=1)
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
output_X = scalar.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(output_X, y, test_size=0.2,random_state = 1)
```

```python
model = Sequential()
model.add(keras.Input(shape=(15,)))
model.add(Dense(7, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(X_train,y_train, epochs=2)
```

```python
model.layers[1].get_weights()
```

```python
y_log = model.predict(X_test)
```

```python
y_pred= np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
